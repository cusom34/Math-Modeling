```python
import numpy as np

A = np.zeros([100,100])
B = np.linspace(1,100,100)

for i in range(100):
     for j in range(100):
          if i == j: A[i,j] = 10
          if i<=95: 
               A[i,i+1] = 1
               A[i,i+2] = 1
               A[i,i+3] = 1
               A[i,i+4] = 1
          if i>=4:
               A[i,i-1] = 1
               A[i,i-2] = 1
               A[i,i-3] = 1
               A[i,i-4] = 1
          if (i <=3 and j<i) or (i >=96 and j>i): A[i,j] = 1
A[97,96] = 10 
A[98,96] = 10
A[99,96] = 10

#print(A[9])
#print(B)
```

# Функции

```python
def gauss_meth(a,b):
    n = len(b)
    a_loc = np.copy(a)
    result = np.copy(b)
    
    for k in range(0,n-1):
        for i in range(k+1,n):
            if a_loc[i,k] != 0.0:
                lam = a_loc[i,k]/a_loc[k,k]
                a_loc[i,k+1:n] = a_loc[i,k+1:n] - lam*a_loc[k,k+1:n]
                result[i] = result[i] - lam*result[k]
        
    for k in range(n-1,-1,-1):
        result[k] = (result[k] - np.dot(a_loc[k,k+1:n],result[k+1:n]))/a_loc[k,k]
    
    return result

def gauss_seidel_meth(A, b, eps):
    n = len(A)
    result = np.zeros_like(b)
    iterations = 0

    while True:
        old_res = np.copy(result)
        for i in range(n):
            temp1 = np.dot(A[i, :i], result[:i])
            temp2 = np.dot(A[i, i+1:], result[i+1:])
            result[i] = (b[i] - temp1 - temp2) / A[i, i]
        iterations +=1

        if np.sqrt(np.dot(result - old_res, result - old_res)) < eps:
            return result, iterations

def evkl_norma_matrix(a):
    s = 0
    for i in range(len(a)):
        for j in range(len(a[0])):   
            s += a[i,j]**2
    result = s**0.5
    return result 

def evkl_norma_vector(a):
    s = 0
    for i in range(len(a)):
        s += a[i]**2
    result = s**0.5
    return result

def matr_difference(a,b):
    result = []
    if len(a) == len(b):
        result += [a[i] - b[i] for i in range(len(a))]
        return result
    else: 
        print('Error: lists have different dimensions')
        return -1

def invert_matrix(a):
    a_loc = np.copy(a)
    result = np.eye(len(a_loc),len(a_loc))
    for fd in range(len(a_loc)):
        fd_scaler = 1.0 / a_loc[fd][fd]
        for j in range(len(a_loc)):
            a_loc[fd][j] *= fd_scaler
            result[fd][j] *= fd_scaler
        
        for i in list(range(len(a_loc)))[0:fd] + list(range(len(a_loc)))[fd+1:]:
            cr_scaler = a_loc[i][fd]
            
            for j in range(len(a_loc)):
                a_loc[i][j] = a_loc[i][j] - cr_scaler * a_loc[fd][j]
                result[i][j] = result[i][j] - cr_scaler * result[fd][j]
    return result
```

## Лямбда максималльная, минимальная и число обусловленности

```python
lam = np.linalg.eig(A)[0]
print('Максимальная лямбда: ', abs(max(lam)))
print('Минимальная лямбда: ', abs(min(lam)), '\n')


A_inv = invert_matrix(A)
#A_inv = np.linalg.inv(A)
print('Евклидова норма матрицы: ',evkl_norma_matrix(A))
print('Евклидова норма обратной матрицы: ',evkl_norma_matrix(A_inv))
print('Число обусловленности матрицы: ',evkl_norma_matrix(A)*evkl_norma_matrix(A_inv))
```

# Метод Гаусса

```python
x1 = gauss_meth(A, B)
print('Решение системы СЛАУ: \n', x1)
#print(np.linalg.solve(A, B)) #для сравнения
```

## Норма вектора невязки

```python
r = matr_difference(np.dot(A,x1),B) #вектор невязки
print('Вектор невязки: \n',r)
print('Норма вектора невязки: ',evkl_norma_vector(r)) #норма вектора невязки
```

# Метод Гаусса-Зейделя

```python
error = 10**(-7)  #точность вычисления
x2, iters = gauss_seidel_meth(A, B, error)
print('Решение системы СЛАУ: \n', x2)
print('Кол-во итераций: ', iters)
#print(np.linalg.solve(A, B)) #для сравнения
```
