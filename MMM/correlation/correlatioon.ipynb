```python
import matplotlib.pyplot as plt
import random

def k(list1, list2):
    correlation = []
    for step in range (len(list1) - len(list2)):
        s = 0
        for i in range(len(list2)):
            s += list1[i+step]*list2[i]
        correlation += [s]
    return correlation

source_signal = []
move_signal = []
mu = 0
sigma = 1

for i in range(999):
    if i <= 449 or i >= 549:
        source_signal += [0 + random.gauss(mu, sigma)]
    else:
        source_signal += [1 + random.gauss(mu, sigma)]

move_signal = source_signal[499 - 75:499 + 75].copy()
print(len(move_signal))
```

```python
axis1 = range(len(source_signal))
plt.plot(axis1,source_signal)

axis2 = range(len(move_signal))
plt.plot(axis2,move_signal)

plt.grid()
plt.xticks(range(min(axis1),max(axis1)+3,100))
plt.title('Искомый сигнал. Sigma = ' + str(sigma))
plt.savefig('source_signal_10')
```

```python
axis = range(len(source_signal) - len(move_signal))
plt.figure(figsize=(14, 4))
plt.plot(axis,k(source_signal, move_signal))
plt.grid()
plt.xticks(range(min(axis),max(axis)+2,25))
plt.title('АКФ. Sigma = ' + str(sigma))
plt.savefig('АКФ_10')
```
