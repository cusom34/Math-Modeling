```python
import numpy as np
import  matplotlib.pyplot as plt

def f(x): #
    if (x>=0):
        return x**(1/3)
    else: return -(abs(x)**(1/3))

def df(x,h): # 
    return (f(x + h) - f(x))/h 

def Df(x):   #. 
        return ((1/3) * 1/(abs(x)**(2/3)))
    
h = 0.001
a = -10
b = 10
x = []
y = []
x2 = []
y2 = []
x3 = []
y3 = []
x4 = []
y4 = []
x5 = []
y5 = []

for i in np.arange(a,b,0.1):
    y+= [f(i)]
    x+=[i]
plt.title('')
plt.plot(x,y)
plt.show()


for i in np.arange(a,b,h):
    y2+= [df(i,h)]
    x2+=[i]
plt.title(' ')
plt.plot(x2,y2)
plt.show()


for i in np.arange(a,b,h):
    y3+= [Df(i)]
    x3+= [i]
plt.title(' ')
plt.plot(x3,y3)
plt.show()    

for i in np.arange(a,b,h):
    y4+= [abs(df(i,h) - Df(i))]
    x4+= [i]
plt.title('')
plt.plot(x4,y4)
plt.show()    

E = 10**(-16)
M = (2/9)* 1/(abs(0.001)**(5/3))
Mo = f(10) 
Ho= 2*(Mo*E/M)**0.5
err =  2*(Mo*E*M)**0.5

for i in np.arange(Ho/10,10*Ho,Ho/10000): #   h,  
    y5+= [0.5*M*i + 2*Mo*E/i]
    x5+= [i]
plt.title('err(h)')
plt.plot(x5,y5)
plt.show()  

print ('Ho = ',Ho)
print ('err = ',err)
```
