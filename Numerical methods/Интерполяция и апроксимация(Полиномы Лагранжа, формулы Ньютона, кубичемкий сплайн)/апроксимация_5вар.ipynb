# Номер 3 (1-я интерполяционная формула Ньютона)



```python
def  a_coef(x, y):

    n = len(y)
    coef = np.zeros([n, n])
    coef[:,0] = y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
                  
    return coef
```

```python
def newton_poly(coef, x_data, x):
    
    n = len(x_data) - 1 
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p
    return p
```

## Для любого столбца

```python
x=[]
y=[]
a=7     #первый год     
b=16    #последний год
c=9     #столбец(месяц) 
for i in range(a, b+1):
    if mass[i][c]!=999.9:
        x+=[mass[i][0]]
        y+=[mass[i][c]]
print(c)
print(x)
print(y)
```

```python
x1=np.arange(min(x),max(x)+0.1, 0.1)
y1=newton_poly(a_coef(x,y)[0, :],x,x1)
```

## 2 задание (тригоном. полином)


```python
from numpy import linalg

def trig(i, a):
    l=1
    
    for j in range(len(y)):
        if i!=j:
            l*=sin((a-x[j])/2)/sin((x[i]-x[j])/2)
    return l

def Tr(x):
    m=[y[i]*trig(i, x) for i in range(len(y))]
    s=0
    for elem in m:
        s+=elem
    return s

def P(point):
    n=len(x)
    a = [0]
    b = [0]
    for k in range(1,int(n/2.)):
        s1 = 0
        s2 = 0
        a0 = 0
        for i in range (len(y)):
            a0 += 1/len(y)*y[i]
            s1 += 2/len(y)*y[i] * cos(2*np.pi*k*i/n)
            s2 += 2/len(y)*y[i] * sin(2*np.pi*k*i/n)
        a += [s1]
        b += [s2]   
    
    bj=[a[k]*cos(2*np.pi*k*(-x[1]+point)/n)  for k in range(1, int(n/2.))]
    aj=[b[k]*sin(2*np.pi*k*(-x[1]+point)/n)  for k in range(1, int(n/2.))]
    s=a0

    for elem in bj:
        s+=elem
    for elem in aj:
        s+=elem
    return s
```

```python
x2=np.arange(min(x),max(x)+0.1, 0.1)
x3=np.arange(min(x)+1,max(x)+0.1, 0.1)
y2 = Tr(x2) #хороша на маленьких промежутках
y3 = P(x3) #хороша на больших промежутках
```
